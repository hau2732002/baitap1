```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, color
from skimage.filters import threshold_otsu,try_all_threshold,threshold_local
rocket =data.rocket()
gray_scaled_rocket=color.rgb2gray(rocket)
plt.imshow(rocket)
```

```python
plt.imshow(gray_scaled_rocket)
```

```python

rocket_vertical_flip = np.flipud (rocket)
plt.imshow(rocket_vertical_flip)
```

```python

red = rocket[:, :, 0]
plt.hist(red.ravel(),  bins=256)
plt.show()
```

```python
thresh = threshold_otsu(gray_scaled_rocket)
binary= gray_scaled_rocket > thresh
plt.imshow(binary)
```

```python
fig,ax=try_all_threshold(gray_scaled_rocket,verbose=False)
plt.show()
```
